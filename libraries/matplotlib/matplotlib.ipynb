```python
import matplotlib.pyplot as plt

years = [2019, 2020, 2021, 2022, 2023]
revenue = [20, 25, 22, 30, 35]

plt.plot(years, revenue, marker='o', color='green')

plt.title("Company Revenue Over Years")
plt.xlabel("Year")
plt.ylabel("Revenue (in Crores)")

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [40, 50, 55, 60, 65, 70, 85, 90]

plt.scatter(study_hours, marks, color='purple', s=50, marker='x')

plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt 

age = [22, 25, 30, 35, 40, 45, 50, 55]
salary = [25000, 30000, 40000, 50000, 55000, 60000, 65000, 70000]

plt.scatter(age, salary, marker='o', s=100, color='red')
plt.xlabel("Age")
plt.ylabel("salary")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

ages = [22, 25, 27, 28, 30, 30, 31, 32, 35, 36, 38, 40, 41, 45, 50, 52, 53, 54, 60]

plt.hist(ages, bins=6, color='orange', edgecolor='black')

plt.title("Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Number of People")
plt.grid(axis='y')
plt.show()
```
