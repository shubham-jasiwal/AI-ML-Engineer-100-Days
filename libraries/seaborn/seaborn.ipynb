```python
import seaborn as sns
import matplotlib.pyplot as plt

products = ["Mobile", "Laptop", "Tablet", "Monitor", "Speaker"]
sales = [100, 150, 120, 80, 130]

import pandas as pd
df = pd.DataFrame({
    "Product": products,
    "Sales": sales
})

sns.barplot(x="Product", y="Sales", data=df, palette="Blues")

plt.title("Sales by Product")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")  

sns.countplot(x="class", data=titanic, palette="pastel")
plt.title("Passenger Class Count")
plt.show()
```

```python
data = {
    "Language": ["Python", "Java", "C++", "JavaScript", "Go"],
    "Users": [100, 80, 60, 90, 30]
}

pdata = pd.DataFrame(data)
sns.barplot(x='Language', y='Users', palette='Blues',data=pdata)
plt.title("Language wise users")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")

numeric_data = titanic.select_dtypes(include='number')

corr = numeric_data.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

selected = titanic[["age", "fare", "pclass", "survived"]]

sns.pairplot(selected, hue="survived", palette="husl")
plt.suptitle("Pairplot: Age, Fare, Class vs Survival", y=1.02)
plt.show()
```
